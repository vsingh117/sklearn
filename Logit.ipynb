```python
# Import required libraries
import pandas as pd
import numpy as np
import sklearn.model_selection
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Read data from input file
inp_data = pd.read_csv("./Data/StudentRecords.csv")
inp_data.head()
```

```python
inp_data['honcomp'] = np.where(inp_data['write']>=60,1,0)
```

```python
inp_data = inp_data[['female', 'read', 'science','honcomp']]
inp_data.head()
```

```python
# Split the data into training and test dataet
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inp_data.drop('honcomp',axis=1), 
                                                    inp_data['honcomp'], test_size=0.30, 
                                                    random_state=101)
```

```python
# Add column 
X_train = sm.add_constant(X_train)
logit_mod = sm.Logit(y_train, X_train)
logit_res = logit_mod.fit()
print(logit_res.summary())
```

```python
X_test = sm.add_constant(X_test)
test_pred = logit_res.predict(X_test)
print('Predicted values for month ::',test_pred)
from sklearn.metrics import mean_squared_error
print('Actual values for month ::\n',y_test)
print('RMSE test values = ', np.sqrt(mean_squared_error(test_pred,y_test)))
```
